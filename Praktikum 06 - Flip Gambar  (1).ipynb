# Flip Gambar

Import dependency

```python
import numpy as np # Library untuk mengolah data dalam bentuk array atau matriks
import imageio # Library untuk membaca, menulis, dan memproses berbagai jenis format file gambar
import matplotlib.pyplot as plt # Library untuk membuat grafik dan visualisasi data dalam bentuk grafik
```

Membaca gambar

```python
img = imageio.imread("ai.jpg") # Membaca gambar
```

Mendapatkan resolusi dan type dari gambar

```python
img_height = img.shape[0] # Mengambil ketinggian (height) gambar dari dimensi array "img" menggunakan indeks ke-0. 
img_width = img.shape[1] # Mengambil lebar (width) gambar dari dimensi array "img" menggunakan indeks ke-1
img_channel = img.shape[2] # Mengambil jumlah channel (warna) gambar dari dimensi array "img" menggunakan indeks ke-2
img_type = img.dtype # Mengambil tipe data gambar menggunakan metode "dtype" dari NumPy
```

Membuat variabel dengan resolusi dan tipe yang sama seperti gambar

```python
img_flip_horizontal = np.zeros(img.shape, img_type) # Membuat array kosong untuk menyimpan gambar yang sudah diflip secara horizontal
img_flip_vertical = np.zeros(img.shape, img_type) # Membuat array kosong untuk menyimpan gambar yang sudah diflip secara vertikal
```

Membalik gambar secara horizontal

```python
for y in range(0, img_height): # Looping untuk setiap piksel pada dimensi vertikal
    for x in range(0, img_width):  # Looping untuk setiap piksel pada dimensi horizontal
        for c in range(0, img_channel): # Looping untuk setiap channel warna
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c] # Melakukan flipping horizontal pada gambar, dan piksel pada kolom x akan diganti dengan piksel pada kolom (img_width-1-x)
```

Membalik gambar secara vertical

```python
for y in range(0, img_height): # Looping untuk setiap piksel pada dimensi vertikal
    for x in range(0, img_width): # Looping untuk setiap piksel pada dimensi horizontal
        for c in range(0, img_channel): # Looping untuk setiap channel warna
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c] # Melakukan flipping vertical pada gambar, dan piksel pada kolom x akan diganti dengan piksel pada kolom (img_height-1-y)
```

Menampilkan hasil balik gambar

```python
plt.imshow(img_flip_horizontal) # Menampilkan gambar yang telah diflip secara horizontal
plt.title("Flip Horizontal") # Memberi judul pada gambar yang ditampilkan
plt.show() # Menampilkan gambar ke layar
plt.imshow(img_flip_vertical) # Menampilkan gambar yang telah diflip secara vertikal
plt.title("Flip Vertical") # Memberi judul pada gambar yang ditampilkan
plt.show() # Menampilkan gambar ke layar
```
